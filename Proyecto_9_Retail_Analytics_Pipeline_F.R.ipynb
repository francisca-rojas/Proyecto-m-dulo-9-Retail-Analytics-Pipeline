## Las 5V del Big Data

El concepto de Big Data se caracteriza por cinco dimensiones fundamentales, conocidas como las 5V: volumen, velocidad, variedad, veracidad y valor. Estas dimensiones permiten comprender los desafíos y oportunidades que surgen al trabajar con grandes cantidades de datos en entornos reales, como el caso de RetailMax.

En primer lugar, el **volumen** hace referencia a la enorme cantidad de datos generados constantemente. En empresas de comercio electrónico, como RetailMax, esto implica millones de transacciones diarias, historiales de compra y registros de navegación. Este gran volumen exige tecnologías escalables, como Apache Spark, que permitan procesar datos de manera distribuida y eficiente, para manejar dicha cantidad de información sin afectar el desempeño.

La **velocidad** se relaciona con la rapidez con la que los datos se generan y deben ser procesados. En plataformas digitales, los datos fluyen en tiempo real, por lo que es fundamental contar con sistemas capaces de procesar información de forma casi inmediata para tomar decisiones oportunas, como recomendaciones de productos o detección de fraudes.

La **variedad** hace referencia a los distintos tipos de datos disponibles. En este contexto, RetailMax maneja datos estructurados (como tablas de ventas), semiestructurados (logs de navegación) y no estructurados (reseñas de usuarios). Esta diversidad requiere herramientas flexibles que puedan integrar y procesar múltiples formatos de datos.

La **veracidad** se refiere a la calidad y confiabilidad de los datos. En proyectos reales, es común encontrar datos incompletos, inconsistentes o con errores. Por ello, es necesario implementar procesos de limpieza y validación que aseguren que los análisis y modelos generados sean confiables.

Finalmente, el **valor** representa el objetivo principal del uso de Big Data: transformar los datos en información útil para la toma de decisiones. En el caso de RetailMax, esto se traduce en obtener insights sobre el comportamiento de los clientes, optimizar estrategias de marketing y mejorar la experiencia del usuario.

En conjunto, estas cinco dimensiones evidencian la necesidad de utilizar tecnologías y enfoques especializados para gestionar y analizar grandes volúmenes de datos, permitiendo a las organizaciones obtener ventajas competitivas en un entorno cada vez más orientado a los datos.

### Transformaciones de limpieza y estructurado de RDD

```python
#Obtener el primer elemento del RDD, en este caso es la fila de encabezado
header = retail_rdd.first()

header
```

```python
#Filtrar para quitar la fila de encabezado
rdd_data = retail_rdd.filter(lambda x: x != header)

rdd_data.take(3)
```

```python
# Separar cada línea del archivo de texto en una lista de valores
rdd_split = rdd_data.map(lambda x: x.split(";"))

rdd_split.take(3)
```

### Crear Pair RDD

Para analizar el comportamiento por cliente (CustomerID, TotalCompra).

```python
# Crear Pair RDD: (CustomerID, TotalCompra)
pair_rdd = rdd_split.map(lambda x: (x[6], float(x[3]) * float(x[5])))

pair_rdd.take(5)
```

```python
# Extraer montos totales
rdd_total_amounts = pair_rdd.map(lambda x: (x[1]))

rdd_total_amounts.take(5)
```

### Transformaciones convenientes

```python
# Extraer productos
rdd_productos = rdd_split.map(lambda x: x[2])

rdd_productos.take(5)
```

```python
# Filtra para eliminar devoluciones (negativos)
rdd_filtered = rdd_split.filter(lambda x: float(x[3]) > 0)

cantidad_ventas = rdd_split.count()
cantidad_sin_devoluciones = rdd_filtered.count()

print("Cantida de devoluciones:")
cantidad_ventas - cantidad_sin_devoluciones
```

```python
# Extraer clientes únicos (Elimina duplicados)
rdd_customers = rdd_split.map(lambda x: x[6]).distinct()

# Cantidad total de clientes únicos
cantidad_clientes = rdd_customers.count()

cantidad_clientes
```

```python
# Ordena el RDD por Cantidad de productos
rdd_sorted = rdd_split.sortBy(lambda x: x[3], ascending=False)

rdd_sorted.take(3)
```

### Métricas de ingresos

- Identificación de productos o períodos más rentables
- Optimización de precios y promociones

```python
# Valor Promedio del Pedido (AOV)
print("\nValor Promedio del Pedido (AOV):")
spark.sql("""
    SELECT
        ROUND(AVG(OrderTotal), 2) AS AverageOrderValue
    FROM (
        SELECT
            InvoiceNo,
            SUM(Quantity * UnitPrice) AS OrderTotal
        FROM retail_data
        WHERE Quantity > 0
        GROUP BY InvoiceNo
    )
""").show()
```

```python
# Productos Top por Ingreso
print("\nProductos Top por Ingreso:")
spark.sql("""
    SELECT
        Description,
        ROUND(SUM(Quantity * UnitPrice), 2) AS TotalRevenue
    FROM retail_data
    WHERE Quantity > 0
    GROUP BY Description
    ORDER BY TotalRevenue DESC
    LIMIT 10
""").show(truncate=False)
```

### Métricas sobre clientes

- Se evalúa comportamiento de clientes.
- Identificación de clientes clave.
- Base para segmentación y fidelización.

```python
# Clientes Top por Gasto
print("\nClientes Top por Gasto:")
spark.sql("""
    SELECT
        CustomerID,
        ROUND(SUM(Quantity * UnitPrice), 2) AS TotalSpend
    FROM retail_data
    WHERE Quantity > 0 AND CustomerID IS NOT NULL AND CustomerID != ''
    GROUP BY CustomerID
    ORDER BY TotalSpend DESC
    LIMIT 10
""").show()
```

```python
print("\nTabla de Clientes con Cantidad de Compras, País y Total Gastado:")
customer_metrics_df = spark.sql("""
    SELECT
        rd.CustomerID,
        COUNT(DISTINCT rd.InvoiceNo) AS TotalUniquePurchases,
        rd.Country,
        ROUND(SUM(rd.Quantity * rd.UnitPrice), 2) AS TotalAmountSpent
    FROM retail_data AS rd
    WHERE rd.Quantity > 0
      AND rd.CustomerID IS NOT NULL
      AND rd.CustomerID != ''
    GROUP BY rd.CustomerID, rd.Country
""")

customer_metrics_df.show()
```

Considerar que solo se toman las compras realizadas, no las devoluciones hechas, si es que las hubo.

### Estrategia para el Tratamiento de Devoluciones y Clientes No Identificados

Para el análisis de ventas y el perfilado de clientes, es fundamental distinguir entre transacciones de compra y devoluciones, así como asegurar que los clientes estén correctamente identificados. Por lo que se hará lo siguiente:
1.  **Separar Ventas y Devoluciones**: Crearemos dos DataFrames separados:
    *   `df_sales`: Contendrá solo las transacciones donde `Quantity` > 0 (compras).
    *   `df_returns`: Contendrá las transacciones donde `Quantity` < 0 (devoluciones).
2.  **Excluir Clientes No Identificados de Análisis de Compras**: Para cualquier análisis que requiera un `CustomerID` (como el perfilado de clientes o el cálculo de métricas por cliente), filtraremos `df_sales` para incluir solo aquellas filas donde `CustomerID` no sea nulo o una cadena vacía. Las transacciones sin `CustomerID` son valiosas para el análisis de productos o ventas totales, pero no para el análisis de clientes individuales.
3.  **Análisis de Devoluciones**: El DataFrame `df_returns` se utilizará para analizar patrones de devolución, clientes con comportamientos de devolución inusuales.

```python
# Separar ventas (Quantity > 0) de devoluciones (Quantity < 0)
df_sales = df_from_parquet.filter(col("Quantity") > 0)
df_returns = df_from_parquet.filter(col("Quantity") < 0)

# Para análisis de clientes, excluimos transacciones sin CustomerID de las ventas
df_sales_customers = df_sales.filter(col("CustomerID").isNotNull() & (col("CustomerID") != ""))

print("Número total de registros de ventas (Quantity > 0):", df_sales.count())
print("Número total de registros de devoluciones (Quantity < 0):", df_returns.count())
print("Número de registros de ventas con CustomerID identificado:", df_sales_customers.count())

print("\nPrimeras 5 filas de df_sales_customers:")
df_sales_customers.show(5)
```

### Preparación de variables para modelos

- Crear features de clientes (Recency), y agregar a datos de clientes
- Encontrar devoluciones y agregar a datos de clientes (TotalReturnedQuantity)
- Calcular monto promedio de compra por cliente y agregar a datos del cliente (AvgOrderValue)
- Se encontraron outliers, por lo cual se decidió aplicar escalado con log para las variables monetarias.
- Definir variable objetivo para regresión logística (Cliente de alto valor)

```python
# Inicializar customer_features_df como base de métricas de clientes
customer_features_df = customer_from_parquet

# Calcular Recency (días desde la última compra)
# Encontrar la fecha de la última compra para cada CustomerID
last_purchase_date_df = df_sales_customers.groupBy("CustomerID") \
    .agg(max("InvoiceDate").alias("LastPurchaseDate"))

# Encontrar la fecha más reciente en todo el dataset de ventas (para calcular Recency)
max_overall_date = df_sales_customers.agg(max("InvoiceDate")).collect()[0][0]
max_overall_date_lit = lit(max_overall_date.strftime("%Y-%m-%d")).cast("date")

# Calcular Recency
recency_df = last_purchase_date_df.withColumn(
    "Recency",
    datediff(max_overall_date_lit, to_date(col("LastPurchaseDate")))
).select("CustomerID", "Recency")

print("Fechas de última compra y Recencia calculadas:")
recency_df.show(5)

# Unir Recency al DataFrame principal de métricas de clientes
customer_features_df = customer_features_df.join(recency_df, on="CustomerID", how="left")

# Calcular el número total de artículos devueltos por cada CustomerID
returns_by_customer_df = df_returns.groupBy("CustomerID") \
    .agg(sum(abs(col("Quantity"))).alias("TotalReturnedQuantity"))

print("Cantidad total de artículos devueltos por cliente (primeras 5 filas):")
returns_by_customer_df.show(5)

# Unir las métricas de devoluciones al DataFrame principal de clientes
customer_features_df = customer_features_df.join(returns_by_customer_df, on="CustomerID", how="left")

# Rellenar los valores nulos en 'TotalReturnedQuantity' con 0, ya que un nulo aquí significa que el cliente no realizó devoluciones
customer_features_df = customer_features_df.na.fill(0, subset=["TotalReturnedQuantity"])
```

```python
# Calcular el Total de Cantidad * Precio Unitario por InvoiceNo (Total de cada pedido)
order_totals = df_sales_customers.withColumn("OrderTotal", col("Quantity") * col("UnitPrice"))

# Calcular el Total de Gasto y el Conteo de Pedidos por CustomerID
customer_order_summary = order_totals.groupBy("CustomerID", "InvoiceNo")\
    .agg(sum("OrderTotal").alias("TotalSpendPerOrder"))

# Ahora, calcular el promedio de TotalSpendPerOrder por CustomerID
avg_order_value_df = customer_order_summary.groupBy("CustomerID")\
    .agg(avg("TotalSpendPerOrder").alias("AvgOrderValue"))

print("Valor promedio del pedido por cliente:")
avg_order_value_df.show(5)

# Unir AvgOrderValue al DataFrame customer_features_df
customer_features_df = customer_features_df.join(avg_order_value_df, on="CustomerID", how="left")

print("DataFrame de clientes con AvgOrderValue:")
customer_features_df.show(5)
```

### Definir variable objetivo para regresión logística: Cliente de alto valor

```python
# Calcular percentiles para TotalAmountSpent
quantiles = customer_features_df.agg(
    percentile_approx("TotalAmountSpent", 0.75).alias("amount_quantile")
).collect()[0]

amount_threshold = quantiles["amount_quantile"]

print(f"Umbral para TotalAmountSpent (75%): {amount_threshold}")

# Crear la variable objetivo is_high_value usando solo TotalAmountSpent
customer_features_df = customer_features_df.withColumn(
    "is_high_value",
    when(
        (col("TotalAmountSpent") >= amount_threshold),
        1
    ).otherwise(0)
)

print("DataFrame con la variable is_high_value:")
customer_features_df.show(5)

print("Distribución de la variable is_high_value:")
customer_features_df.groupBy("is_high_value").count().show()
```

### Preprocesamiento de features
- Codificación para variables categóricas ("Country")
- Escalado para variables numéricas ("Recency", "TotalUniquePurchases", "TotalReturnedQuantity", "LogAvgOrderValue", previamente había sido considerada "LogTotalAmountSpent", pero fue descartada por sesgar el modelo al ser la variable que define alto valor)
- Creación de VectorAssembler

```python
# Eliminar filas con CustomerID nulo si las hubiera
customer_features_df = customer_features_df.filter(col("CustomerID").isNotNull() & (col("CustomerID") != ""))

# StringIndexer para 'Country'
indexer = StringIndexer(inputCol="Country", outputCol="CountryIndex")
indexed_df = indexer.fit(customer_features_df).transform(customer_features_df)

print("DataFrame con 'CountryIndex':")
indexed_df.select("CustomerID", "Country", "CountryIndex").show(5)

# OneHotEncoder para 'CountryIndex'
# Esto convierte los índices numéricos en vectores one-hot codificados.
encoder = OneHotEncoder(inputCol="CountryIndex", outputCol="CountryVec")
encoded_df = encoder.fit(indexed_df).transform(indexed_df)

print("DataFrame con 'CountryVec':")
encoded_df.select("CustomerID", "Country", "CountryIndex", "CountryVec").show(5, truncate=False)
```

```python
# Lista de columnas numéricas
numeric_cols = ["Recency", "TotalUniquePurchases", "TotalReturnedQuantity", "LogAvgOrderValue"]

# Verificar la presencia de NaN o valores infinitos en las columnas numéricas que usaremos como features
print("Verificando NaN o valores infinitos en columnas numéricas antes de VectorAssembler:")
for c in numeric_cols:
    # Verificar NaN
    nan_count = encoded_df.filter(isnan(col(c))).count()
    # Verificar valores infinitos (solo relevante para float/double)
    inf_count = encoded_df.filter(col(c).cast("string").like("%Infinity%")).count()
    if nan_count > 0 or inf_count > 0:
        print(f"  Columna '{c}': NaN={nan_count}, Infinitos={inf_count}")

# Rellenar cualquier NaN en 'LogAvgOrderValue' si es necesario.
# Si no hay NaN o Inf, este paso no tendrá efecto.
avg_log_avg_order_value = encoded_df.select(avg(col("LogAvgOrderValue"))).collect()[0][0]
encoded_df = encoded_df.na.fill(avg_log_avg_order_value, subset=["LogAvgOrderValue"])

# Las columnas de características ya están definidas correctamente para usar las versiones log-transformadas.
feature_cols = ["Recency", "TotalUniquePurchases", "TotalReturnedQuantity", "LogAvgOrderValue", "CountryVec"]

# VectorAssembler
# Combina todas las columnas de características en una única columna de vector densa
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
features_df = assembler.transform(encoded_df)

# Mostrar el DataFrame final con las características y la variable objetivo
print("DataFrame final con características y variable objetivo (primeras 5 filas):")
features_df.select("CustomerID", "features", "is_high_value").show(5, truncate=False)
```

```python
# Inicializar StandardScaler
# withMean=True y withStd=True para estandarizar las características a una media de 0 y desviación estándar de 1
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withMean=True, withStd=True)

# Aplicar el escalador en el DataFrame de características
scaler_model = scaler.fit(features_df)

# Transformar el DataFrame para obtener las características escaladas
scaled_features_df = scaler_model.transform(features_df)

print("DataFrame con características escaladas:")
scaled_features_df.select("CustomerID", "features", "scaledFeatures", "is_high_value").show(5, truncate=False)
```

### Entrenar modelo de Regresión Logística

- División de datos de forma estratificada
- Entrenar modelo
- Hacer predicción
- Evaluar modelo con curva ROC, matriz de confusión y accuracy.
- Optimización del modelo
- Elección del modelo a utilizar

```python
# Separar el DataFrame por la variable objetivo 'is_high_value'
high_value_customers = scaled_features_df.filter(scaled_features_df.is_high_value == 1)
non_high_value_customers = scaled_features_df.filter(scaled_features_df.is_high_value == 0)

# Realizar randomSplit estratificado en cada grupo
(trainingData_high_value, testData_high_value) = high_value_customers.randomSplit([0.7, 0.3], seed=42)
(trainingData_non_high_value, testData_non_high_value) = non_high_value_customers.randomSplit([0.7, 0.3], seed=42)

# Unir los conjuntos de entrenamiento y prueba estratificados
trainingData_stratified = trainingData_high_value.unionAll(trainingData_non_high_value)
testData_stratified = testData_high_value.unionAll(testData_non_high_value)

print("Número de registros para entrenamiento (estratificado):", trainingData_stratified.count())
print("Número de registros para prueba (estratificado):", testData_stratified.count())

print("\nDistribución de 'is_high_value' en el conjunto de entrenamiento estratificado:")
trainingData_stratified.groupBy("is_high_value").count().show()

print("\nDistribución de 'is_high_value' en el conjunto de prueba estratificado:")
testData_stratified.groupBy("is_high_value").count().show()

# Ahora, usaremos trainingData_stratified y testData_stratified para entrenar y evaluar el modelo.
```

Ahora que tenemos los datos divididos de forma estratificada, podemos proceder a entrenar y evaluar el modelo de Regresión Logística utilizando `trainingData_stratified` y `testData_stratified`. Esto asegurará que ambos conjuntos tengan una representación proporcional de clientes de alto valor, mejorando la robustez de nuestro modelo y la fiabilidad de nuestras métricas de evaluación.

```python
# Inicializar el modelo de Regresión Logística con los datos estratificados
lrc_stratified = LogisticRegression(labelCol="is_high_value", featuresCol="scaledFeatures", maxIter=10)

# Entrenar el modelo con los datos de entrenamiento estratificados
lr_model_stratified = lrc_stratified.fit(trainingData_stratified)

print("Modelo de Regresión Logística entrenado con datos estratificados.")

# Realizar predicciones en el conjunto de prueba estratificado
predictions_stratified = lr_model_stratified.transform(testData_stratified)

print("Primeras 5 predicciones con datos estratificados:")
predictions_stratified.select("CustomerID", "is_high_value", "prediction", "probability").show(5)

# Evaluar el modelo con datos estratificados
evaluator_stratified = BinaryClassificationEvaluator(labelCol="is_high_value")
auc_stratified = evaluator_stratified.evaluate(predictions_stratified, {evaluator_stratified.metricName: "areaUnderROC"})

print(f"\nÁrea bajo la curva ROC (AUC) para el modelo de Regresión Logística: {auc_stratified}")
```

### Nueva Predicción con `threshold = 0.4(0.3)`

Vamos a generar una nueva columna de predicción aplicando un umbral de 0.4, luego para 0.3 a las probabilidades predichas por el modelo de Regresión Logística. Luego evaluaremos el rendimiento del modelo con este nuevo umbral.

```python
# Definir el nuevo umbral
new_threshold = 0.3

# Define a UDF to extract the probability of the positive class (index 1) from the vector
get_prob_udf = udf(lambda v: float(v[1]), DoubleType())

# Crear una nueva columna de predicción basada en el nuevo umbral
predictions_thresholded = predictions_stratified.withColumn(
    "prediction_0_4",
    when(get_prob_udf(col("probability")) >= new_threshold, 1.0).otherwise(0.0)
)

print(f"Primeras 5 predicciones con threshold={new_threshold}:")
predictions_thresholded.select("CustomerID", "is_high_value", "probability", "prediction_0_4").show(5)

# Obtener las etiquetas verdaderas y las nuevas predicciones
y_true_new = predictions_thresholded.select("is_high_value").toPandas()["is_high_value"]
y_pred_new = predictions_thresholded.select("prediction_0_4").toPandas()["prediction_0_4"]

# Calcular la nueva matriz de confusión
cm_new = confusion_matrix(y_true_new, y_pred_new)
print(f"\nMatriz de Confusión (threshold={new_threshold}):\n", cm_new)

# Calcular la nueva precisión (Accuracy)
accuracy_new = accuracy_score(y_true_new, y_pred_new)
print(f"\nAccuracy (threshold={new_threshold}): {accuracy_new:.4f}")

# Visualizar la nueva Matriz de Confusión como Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm_new, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=["No High Value", "High Value"], yticklabels=["No High Value", "High Value"])
plt.title(f"Matriz de Confusión para Cliente de Alto Valor (Threshold = {new_threshold})")
plt.xlabel("Predicción")
plt.ylabel("Valor Real")
plt.show()
```

Por otro lado, los falsos positivos aumentan de 21 a 29, lo que representa un incremento en el número de clientes clasificados como de alto valor sin serlo. Sin embargo, este aumento es moderado y consistente con el ajuste del umbral.

Aunque se incrementan los falsos positivos, el costo asociado a contactar clientes adicionales es considerablemente menor que el de no identificar clientes valiosos.

### Optimización modelo de regresión logística con ParamGridBuilder

```python
# Crear una instancia del modelo de Regresión Logística
lrc_cv = LogisticRegression(labelCol="is_high_value", featuresCol="scaledFeatures")

# Definir la cuadrícula de parámetros a buscar
paramGrid = (
    ParamGridBuilder()
    .addGrid(lrc_cv.maxIter, [10, 50]) # Número de iteraciones
    .addGrid(lrc_cv.regParam, [0.01, 0.1, 0.5]) # Parámetro de regularización
    .addGrid(lrc_cv.elasticNetParam, [0.0, 0.5, 1.0]) # Mezcla entre L1 y L2 (0=L2, 1=L1)
    .build())

# Crear un evaluador para el modelo de clasificación binaria
evaluator_cv = BinaryClassificationEvaluator(labelCol="is_high_value", rawPredictionCol="rawPrediction", metricName="areaUnderROC")

# Crear el CrossValidator
cv = CrossValidator(
    estimator=lrc_cv, # El modelo a entrenar
    estimatorParamMaps=paramGrid, # La cuadrícula de parámetros
    evaluator=evaluator_cv, # El evaluador para seleccionar el mejor modelo
    numFolds=3, # Número de folds para la validación cruzada
    seed=42
)

# Entrenar el CrossValidator para encontrar el mejor modelo
print("Entrenando CrossValidator para encontrar el mejor modelo...")
cvModel = cv.fit(trainingData_stratified) # Usamos los datos de entrenamiento estratificados

print("CrossValidator ha completado el entrenamiento.")

# Obtener el mejor modelo
best_lr_model = cvModel.bestModel

print("Mejor modelo de Regresión Logística encontrado:")
print(best_lr_model.extractParamMap())

# Realizar predicciones con el mejor modelo en el conjunto de prueba
predictions_best_model = best_lr_model.transform(testData_stratified)

# Evaluar el mejor modelo
auc_best_model = evaluator_cv.evaluate(predictions_best_model)
print(f"\nÁrea bajo la curva ROC (AUC) del mejor modelo de Regresión Logística: {auc_best_model:.4f}")
```

### Matriz de Confusión y Métricas de Evaluación para el Mejor Modelo de Regresión Logística

```python
# Obtener las etiquetas verdaderas y las predicciones del mejor modelo
y_true_best = predictions_best_model.select("is_high_value").toPandas()["is_high_value"]
y_pred_best = predictions_best_model.select("prediction").toPandas()["prediction"]

# Calcular la matriz de confusión
cm_best = confusion_matrix(y_true_best, y_pred_best)
print("Matriz de Confusión para el Mejor Modelo:\n", cm_best)

# Calcular la precisión (Accuracy)
accuracy_best = accuracy_score(y_true_best, y_pred_best)
print(f"\nAccuracy para el Mejor Modelo: {accuracy_best:.4f}")

# Calcular Precision
precision_best = precision_score(y_true_best, y_pred_best)
print(f"Precision para el Mejor Modelo: {precision_best:.4f}")

# Calcular Recall
recall_best = recall_score(y_true_best, y_pred_best)
print(f"Recall para el Mejor Modelo: {recall_best:.4f}")

# Visualizar la Matriz de Confusión como Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm_best, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=["No High Value", "High Value"], yticklabels=["No High Value", "High Value"])
plt.title("Matriz de Confusión para el Mejor Modelo de Cliente de Alto Valor")
plt.xlabel("Predicción")
plt.ylabel("Valor Real")
plt.show()
```

### Segmentación de Clientes con K-Means (Modelo No Supervisado)

```python
# Seleccionaremos las columnas necesarias para K-Means
kmeans_data = scaled_features_df.select("CustomerID", "scaledFeatures")

# Entrenar un modelo K-Means
k = 2
kmeans = KMeans(featuresCol="scaledFeatures", k=k, seed=42, maxIter=20, tol=0.0001, initSteps=5)
kmeans_model = kmeans.fit(kmeans_data)

# Realizar predicciones de clúster en los datos
kmeans_predictions = kmeans_model.transform(kmeans_data)

print(f"Segmentación de Clientes con K-Means (k={k}):")
kmeans_predictions.show(5)

# Evaluar la segmentación utilizando la métrica Silhouette
evaluator = ClusteringEvaluator(featuresCol="scaledFeatures", predictionCol="prediction",
                                metricName="silhouette", distanceMeasure="squaredEuclidean")

silhouette = evaluator.evaluate(kmeans_predictions)
print(f"\nMétrica Silhouette con k={k}: {silhouette}")

# Contar el número de clientes en cada clúster
print("\nDistribución de clientes por clúster:")
kmeans_predictions.groupBy("prediction").count().orderBy("prediction").show()
```

### Optimizar el modelo con evaluación de hiperparámetros

```python
# Definir los rangos de hiperparámetros a probar
k_range = range(2, 6)
maxIter_range = [20, 50]
tol_range = [1e-4, 1e-5, 1e-6]
initSteps_range = [5, 10]

# Variables para almacenar los mejores resultados
best_silhouette = -1
best_k = -1
best_maxIter = -1
best_tol = -1.0
best_initSteps = -1

# Preparar los datos para K-Means (ya escalados)
kmeans_data_for_search = scaled_features_df.select("scaledFeatures")

print("Iniciando búsqueda de hiperparámetros para K-Means...")

for k_val in k_range:
    for mi_val in maxIter_range:
        for tol_val in tol_range:
            for is_val in initSteps_range:
                print(f"Probando k={k_val}, maxIter={mi_val}, tol={tol_val}, initSteps={is_val}")
                # Crear el modelo K-Means con los hiperparámetros actuales
                kmeans = KMeans(
                    featuresCol="scaledFeatures",
                    k=k_val,
                    seed=42,
                    maxIter=mi_val,
                    tol=tol_val,
                    initSteps=is_val
                )

                # Entrenar el modelo
                kmeans_model = kmeans.fit(kmeans_data_for_search)
                predictions = kmeans_model.transform(kmeans_data_for_search)

                # Calcular el coeficiente de silueta
                evaluator = ClusteringEvaluator(
                    featuresCol="scaledFeatures",
                    predictionCol="prediction",
                    metricName="silhouette",
                    distanceMeasure="squaredEuclidean"
                )
                silhouette = evaluator.evaluate(predictions)
                print(f"  -> Silueta: {silhouette:.4f}")

                # Actualizar los mejores resultados si se encuentra una mejor silueta
                if silhouette > best_silhouette:
                    best_silhouette = silhouette
                    best_k = k_val
                    best_maxIter = mi_val
                    best_tol = tol_val
                    best_initSteps = is_val

print("\n--- Resultados de la Búsqueda de Hiperparámetros ---")
print(f"Mejor coeficiente de Silueta: {best_silhouette:.4f}")
print(f"Mejor k: {best_k}")
print(f"Mejor maxIter: {best_maxIter}")
print(f"Mejor tol: {best_tol}")
print(f"Mejor initSteps: {best_initSteps}")
```

### Optimización y selección del modelo de clustering

Se intentó optimizar el modelo de clustering mediante una búsqueda de hiperparámetros, evaluando distintas configuraciones de K-Means en función del coeficiente de silueta. En una primera instancia, se obtuvo un resultado con un valor muy alto de silueta (0.9837), lo que indicaría una excelente separación entre clústeres.

Sin embargo, al analizar la distribución de los datos, se observó que los clústeres no tenían sentido desde el punto de vista de negocio, ya que se generaba una segmentación altamente desbalanceada, donde un clúster concentraba prácticamente la totalidad de los clientes (más de 4.300) y otro contenía solo un caso aislado. Este comportamiento sugiere que el modelo estaba identificando outliers en lugar de generar segmentos representativos, lo que limita completamente su utilidad comercial.

Distribución de clientes por clúster:

Posteriormente, se evaluaron configuraciones con un mayor número de clústeres (k=3 y k=4). No obstante, se volvió a observar la aparición de grupos extremadamente pequeños, lo que refuerza la hipótesis de la presencia de valores atípicos en el dataset y la tendencia del algoritmo a aislarlos para maximizar la métrica de silueta.

En paralelo, se exploraron enfoques complementarios como el método del codo y la reducción de dimensionalidad mediante PCA. Sin embargo, estos métodos no aportaron información concluyente para la selección del número de clústeres en este caso.

Durante el proceso, se identificó que el parámetro `initSteps` tenía un impacto relevante en la formación de los clústeres. Al ajustar este parámetro, se logró obtener un modelo con un coeficiente de silueta menor, pero con una distribución mucho más equilibrada.

Esta configuración permite identificar dos segmentos claros y accionables: un grupo mayoritario de clientes con comportamiento estándar y un grupo minoritario de clientes de alto valor.
